```python
import librosa
```

```python
test_song = '/kaggle/input/gtzan-dataset-music-genre-classification/Data/genres_original/blues/blues.00035.wav'
song, sr = librosa.load(test_song, mono=False)
np.shape(song)
```

The original values for window and hop were hard coded at 2048, 512.
Seems like there may be some potential here for feature engineering / optimization.

[librosa stft documentation](https://librosa.org/doc/latest/generated/librosa.stft.html#librosa.stft)

```python
def calc_stft(audio_file, n_fft=2048, hop_length=512):
    # Short-time Fourier transform (STFT)
    D = np.abs(librosa.stft(audio_file, n_fft = n_fft, hop_length = hop_length))
    
    print('Shape of D object:', np.shape(D))
```

```python
stft = calc_stft(song)

print(f'Sample Rate: {sr}')
print(f'Num Samples {np.shape(song)}')
print(f'Length: {np.shape(song)[0]/sr}')
```
